2.  Distribusi Tahun Pembuatan dengan Harga, Mengetahui apakah mobil keluaran terbaru cenderung lebih mahal (linechart)

#### Statistik Dekriptif

Pola pesebaran data pada harga

```python

skew_price = skew(df['Price'])

print(f"\nSkewness Harga: {skew_price}")
```

Skewness > 0 → Distribusi miring ke kanan (positively skewed). Ini berarti ada lebih banyak nilai kecil dengan beberapa nilai besar yang menjadi outlier.<br>
Skewness < 0 → Distribusi miring ke kiri (negatively skewed). Artinya ada lebih banyak nilai besar dengan beberapa nilai kecil sebagai outlier.<br>
Skewness = 0 → Distribusi simetris seperti distribusi normal<br>

Skewness pada data harga mobil dalam dataset ini memiliki nilai -0.014, yang sangat mendekati nol. Hal ini menunjukkan bahwa distribusi harga hampir simetris, tanpa kemiringan signifikan ke kiri atau kanan. Dengan kata lain, harga mobil tersebar secara merata di sekitar nilai rata-rata, tanpa didominasi oleh harga yang sangat tinggi atau sangat rendah. Dalam distribusi yang memiliki skewness negatif, seharusnya terdapat lebih banyak harga tinggi dengan beberapa outlier harga rendah. Namun, karena nilai skewness dalam dataset ini sangat kecil, dapat disimpulkan bahwa tidak ada bias signifikan terhadap harga mahal atau murah. 

#### Statistik Inferensial

 Harga Mobil Otomatis vs Manual

```python
from scipy.stats import ttest_ind

# Pisahkan data berdasarkan jenis transmisi
auto_price = df[df['Transmission'] == 'Automatic']['Price']
manual_price = df[df['Transmission'] == 'Manual']['Price']

# Lakukan uji t-test independen
statistic, p_value = ttest_ind(auto_price, manual_price, equal_var=False)  # gunakan equal_var=True jika varians homogen

# Tampilkan hasil uji
print("\nHasil Uji T-Test (Harga Mobil Otomatis vs Manual):")
print(f"T-Statistic: {statistic}")
print(f"P-Value: {p_value}")

# Interpretasi hasil uji
alpha = 0.05  # Signifikansi 5%
if p_value < alpha:
    print("Terdapat perbedaan signifikan antara harga mobil otomatis dan manual.")
else:
    print("Tidak terdapat perbedaan signifikan antara harga mobil otomatis dan manual.")
```

```python
from scipy.stats import mannwhitneyu

# Pisahkan data berdasarkan jenis transmisi
auto_price = df[df['Transmission'] == 'Automatic']['Price']
manual_price = df[df['Transmission'] == 'Manual']['Price']

# Lakukan uji Mann-Whitney U
statistic, p_value = mannwhitneyu(auto_price, manual_price)

print("\nHasil Uji Mann-Whitney U (Harga Mobil Otomatis vs Manual):")
print(f"U-Statistic: {statistic}")
print(f"P-Value: {p_value}")

# Interpretasi hasil uji
alpha = 0.05  # Signifikansi 5%
if p_value < alpha:
    print("Terdapat perbedaan signifikan antara harga mobil otomatis dan manual.")
else:
    print("Tidak terdapat perbedaan signifikan antara harga mobil otomatis dan manual.")
```

Analisis statistik menggunakan Uji Mann-Whitney U dilakukan untuk membandingkan harga mobil berdasarkan jenis transmisi, yaitu otomatis dan manual. Uji ini merupakan metode non-parametrik yang digunakan ketika asumsi normalitas tidak dapat dipenuhi, sehingga lebih fleksibel dibandingkan uji t. Dari hasil analisis, diperoleh U-Statistic sebesar 781412.0 dan p-value sebesar 0.9185. Dengan taraf signifikansi α = 0.05, keputusan uji didasarkan pada nilai p-value. Karena p-value jauh lebih besar dari 0.05, maka hipotesis nol (H₀) tidak dapat ditolak, yang berarti tidak terdapat perbedaan signifikan antara harga mobil otomatis dan manual.

 Harga Mobil Baru >2015 vs Lama <=2015

```python
from scipy.stats import mannwhitneyu

# Pisahkan data berdasarkan tahun produksi
new_cars = df[df['Year'] > 2015]['Price']
old_cars = df[df['Year'] <= 2015]['Price']

# Lakukan uji Mann-Whitney U
statistic, p_value = mannwhitneyu(new_cars, old_cars)

print("\nHasil Uji Mann-Whitney U (Harga Mobil Baru >2015 vs Lama ≤ 2015):")
print(f"U-Statistic: {statistic}")
print(f"P-Value: {p_value}")

# Interpretasi hasil uji
alpha = 0.05  # Signifikansi 5%
if p_value < alpha:
    print("Terdapat perbedaan signifikan antara harga mobil baru dan lama.")
else:
    print("Tidak terdapat perbedaan signifikan antara harga mobil baru dan lama.")
```

 Uji Mann-Whitney U bertujuan untuk membandingkan harga mobil baru (tahun produksi > 2015) dengan harga mobil lama (tahun produksi ≤ 2015). Dari hasil analisis, diperoleh nilai U-Statistic sebesar 66828.5 dan p-value sebesar 0.0184. Dengan menggunakan taraf signifikansi α = 0.05, hasil ini menunjukkan bahwa p-value lebih kecil dari α, sehingga hipotesis nol (H₀) ditolak. Artinya, terdapat perbedaan signifikan antara harga mobil baru dan harga mobil lama. Namun, uji Mann-Whitney U hanya mengidentifikasi adanya perbedaan tanpa menentukan arah perubahan harga.
